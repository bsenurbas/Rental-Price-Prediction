```python
# Veri temizleme işlemleri için gerekli fonksiyon
def drop_columns(df, cols):
    '''
    Belirtilen sütunları DataFrame'den siler.
    '''
    for col in cols:
        try:
            df.drop( col,axis=1, inplace=True) # belirtilen sütunları siler
        except Exception as e:
            print(f"Hata: {e}") # hata mesajını yazdırır
```

```python
# Örnek bir URL'den sütunları ayırma
# Bu örnek, URL'yi '/' karakterine göre böler ve 3. sütunu alır.
# Bu, genellikle bir web sayfasının belirli bir bölümünü veya öğesini temsil eder.
s = "https://www.hepsiemlak.com/ankara-mamak-general-zeki-dogan-kiralik/daire/159186-7 "
x = s.split('/')[3] # sütunları / ile ayırır
print(x) # 3. sütunu yazdırır
```

```python
# Yeni bir sütun ekleme
df ['location'] = df['img-link href'].apply(lambda x: x.split('/')[3]) # img-link href sütununu / ile ayırır ve 3. sütunu alır
df.info() # df ile ilgili bilgileri tekrar verir, yeni sütun eklenmiştir
df['city'] = df['location'].str.split('-').str[0] # location sütununu - ile ayırır ve 0. sütunu alır
df['district'] = df['location'].str.split('-').str[1] # location sütununu - ile ayırır ve 1. sütunu alır
df['neighborhood'] = df['location'].str.split('-').str[2:-1] # location sütununu - ile ayırır ve 2. sütundan sonrasını alır
df['neighborhood'] = df['neighborhood'].apply(lambda x: '_'.join(x)) # neighborhood sütununu birleştirir
print(df[['city', 'district', 'neighborhood']]) # yeni sütunları gösterir
```

```python
drop_columns(df, ['location','img-link href']) # belirtilen sütunları siler
df.info() # df ile ilgili bilgileri tekrar verir, yeni sütunlar silinmiştir
print(df[['city','district','neighborhood']]) # benzersiz şehirleri listeler
```

```python
print(df.columns)
print(df['celly'].unique()) # benzersiz sütunları listeler
```

```python
df['celly'] = df['celly'].apply(lambda x: x.replace('Stüdyo', '1+0')) # 'celly' sütunundaki 'Stüdyo' kelimesini '1+0' ile değiştirir
df['celly'] = df['celly'].apply(lambda x: x.replace('\n', '')) # 'celly' sütunundaki '\n' karakterini boşluk ile değiştirir
df['room'] = df['celly'].apply(lambda x: x.split('+')[0]).astype(int) # 'celly' sütununu '+' ile ayırır ve 0. sütunu alır, int tipine çevirir
df['livingroom'] = df['celly'].apply(lambda x: x.split('+')[1]).astype(int) # 'celly' sütununu '+' ile ayırır ve 1. sütunu alır, int tipine çevirir
print(df['room'].unique()) # benzersiz oda sayılarını listeler
print(df['livingroom'].unique()) # benzersiz salon sayılarını listeler
drop_columns(df, ['celly']) # 'celly' sütununu siler
df.info() # df ile ilgili bilgileri tekrar verir, 'celly' sütunu artık yoktur
```

```python
print(df['celly 2'].unique()) # 'celly 2' sütunundaki benzersiz değerleri listeler
```

```python
df['celly 2'] = df['celly 2'].apply(lambda x: x.replace('.', '')) # 'celly 2' sütunundaki '.' karakterini boşluk ile değiştirir
df['area'] = df['celly 2'].apply(lambda x: x.split(' ')[0]).astype(int) # 'celly 2' sütununu ' ' ile ayırır ve 0. sütunu alır, int tipine çevirir
```

```python
df['celly 3'] = df['celly 3'].apply(lambda x: x.replace('Sıfır Bina', '0')) # 'celly 3' sütunundaki 'Sıfır Bina' kelimesini '0' ile değiştirir
df['celly 3'] = df['celly 3'].apply(lambda x: x.replace('\n', '')) # 'celly 3' sütunundaki '\n' karakterini boşluk ile değiştirir
df['age'] = df['celly 3'].apply(lambda x: x.split(' ')[0]).astype(int) # 'celly 3' sütununu ' ' ile ayırır ve 0. sütunu alır, int tipine çevirir
print(df['age'].unique()) # benzersiz yaşları listeler
```

```python
'''search_for = [
    'Kot 2',
    'Kot 1'
    ] 
replace_with = [
    '-2. Kat', 
    '-1. Kat'
    ] # arama ve değiştirme işlemleri için kullanılacak değerler'''
replace_dict = {
    'Kot 2': '-2. Kat',
    'Kot 1': '-1. Kat',
    'Kot 3': '-3. Kat',
    'Yüksek Giriş': '1. Kat',
    'Giriş Katı': '1. Kat',
    'Ara Kat': '3. Kat',
    'Zemin': '0. Kat',
    'En Üst Kat': '5. Kat',
    'Bahçe Katı': '0. Kat',
    'Yarı Bodrum': '0. Kat',
    'Teras Katı': '5. Kat',
    '21 ve üzeri': '21. Kat',
    'Çatı Katı': '5. Kat',
    'Villa Katı': '5. Kat',
    'Bodrum': '0. Kat',
    'Bodrum ve Zemin': '0. Kat',
    'Tripleks': '3. Kat',
    'nan': '2. Kat',
} # arama ve değiştirme işlemleri için kullanılacak değerler
df['celly 4'] = df['celly 4'].replace(list(replace_dict.keys()), list(replace_dict.values())).astype(str) # 'celly 4' sütunundaki değerleri replace_dict ile değiştirir, str tipine çevirir
df['floor'] = df['celly 4'].apply(lambda x: x.split('.')[0]).astype(int)# 'celly 4' sütununu ' ' ile ayırır ve 0. sütunu alır
```

```python
print(df['floor'].unique()) # benzersiz kat sayılarını listeler
drop_columns(df, ['celly 4']) # 'celly 4' sütununu siler
print(df.info()) # df ile ilgili bilgileri tekrar verir, 'celly 4' sütunu artık yoktur  
```
